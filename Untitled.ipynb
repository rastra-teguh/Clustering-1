# Encoding and Checking the EDA

```python
df['category_code'] = df['category_name'].astype('str')
```

```python
df.head(2)
```

```python
df['category_code'].unique()
```

```python
def category_code(x):
    segment = ''
    if  x['category_code'] == 'Movies' :
        segment = 1
    elif x['category_code'] == 'News' :
        segment = 2
    elif x['category_code'] == 'Entertainment' :
        segment = 3
    elif x['category_code'] == 'Sports' :
        segment = 4
    elif x['category_code'] == 'TV Show' :
        segment = 5
    elif x['category_code'] == 'Korea' :
        segment = 6
    elif x['category_code'] == 'Lifestyle' :
        segment = 7
    elif x['category_code'] == 'Vlog' :
        segment = 8
    elif x['category_code'] == 'Horror' :
        segment = 9
    elif x['category_code'] == 'Kids' :
        segment = 10
    elif x['category_code'] == 'Funny' :
        segment = 11
    elif x['category_code'] == 'Education' :
        segment = 12
    elif x['category_code'] == 'Webtoon' :
        segment = 13
    elif x['category_code'] == 'Champions' :
        segment = 14
    else:
        segment = 15
    return segment

df['category_code'] = df.apply(lambda x : category_code(x), axis=1)
df.head(10)
```

```python
df['view'] = df['view_time'].astype(int)
df.head(2)
```

```python
time1_4 = df.view[(df.view <= 4) & (df.view >= 1)]
time5_8 = df.view[(df.view <= 8) & (df.view >= 5)]
time9_12 = df.view[(df.view <= 9) & (df.view >= 12)]
time13_16 = df.view[(df.view <= 13) & (df.view >= 16)]
time17_20 = df.view[(df.view <= 17) & (df.view >= 20)]
time21_0 = df.view[df.view >= 21 & (df.view >= 23)& (df.view == 0)] 

x = ["1-4","5-8","9-12","13-16","17-20","21-0"]
y = [len(time1_4.values),len(time5_8.values),len(time9_12.values),len(time13_16.values),len(time17_20.values),len(time21_0.values)]

plt.figure(figsize=(15,6))
sns.barplot(x=x, y=y, palette="rocket")
plt.title("view time vs category")
plt.xlabel("view time")
plt.ylabel("number of viewer")
plt.show()
```

# Clustering

```python
from sklearn.cluster import KMeans
wcss = []
for k in range(1,11):
    kmeans = KMeans(n_clusters=k, init="k-means++")
    kmeans.fit(new[['category_code','view']])
    wcss.append(kmeans.inertia_)
plt.figure(figsize=(12,6))    
plt.grid()
plt.plot(range(1,11),wcss, linewidth=2, color="red", marker ="8")
plt.xlabel("K Value")
plt.xticks(np.arange(1,11,1))
plt.ylabel("WCSS")
plt.show()
```

```python
km = KMeans(n_clusters=4)
```

```python
clusters = km.fit_predict(new.iloc[:,1:])
```

```python
new["label"] = clusters
```

```python
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(new.view[new.label == 0], new["category_code"][new.label == 0], df["play_duration"][new.label == 0], c='blue', s=60)
ax.scatter(new.view[new.label == 1], new["category_code"][new.label == 1], df["play_duration"][new.label == 1], c='red', s=60)
ax.scatter(new.view[new.label == 2], new["category_code"][new.label == 2], df["play_duration"][new.label == 2], c='green', s=60)
ax.scatter(new.view[new.label == 3], new["category_code"][new.label == 3], df["play_duration"][new.label == 3], c='orange', s=60)

ax.view_init(30, 185)
plt.xlabel("view")
plt.ylabel("categorty_code")
ax.set_zlabel('play_duration')
plt.show()
```
